## Full JaQuad train data analysis into dataframe

```python
import pandas as pd
train_df_list = []
for entry in train_list:
    with open(f"data/train/{entry}", "r") as foo:
        train_sample = json.load(foo)


    train_sample_dict_list = []
    for wiki_article in train_sample["data"]:
        
        for context in wiki_article["paragraphs"]:
            
            for qas in context["qas"]:
                # print(qas)
                for i, ans in enumerate(qas["answers"]):
                    record_dict = {}
                    record_dict["title"] = wiki_article["title"]
                    record_dict["context"] = context["context"]
                    record_dict["question"] = qas["question"]
                    record_dict["id"] = qas["id"]
                    record_dict["answers"] = qas["answers"][i]["text"]
                    record_dict["answer_start"] = qas["answers"][i]["answer_start"]
                    record_dict["answer_type"] = qas["answers"][i]["answer_type"]
                    train_sample_dict_list.append(record_dict)
    
    
    train_sample_df = pd.DataFrame.from_records(train_sample_dict_list)
    train_df_list.append(train_sample_df)

train_df = pd.concat(train_df_list, axis=0)
train_df.shape
```

## Full JaQuad dev data analysis into dataframe

```python
import pandas as pd
dev_df_list = []
for entry in dev_list:
    with open(f"data/dev/{entry}", "r") as foo:
        dev_sample = json.load(foo)


    dev_sample_dict_list = []
    for wiki_article in dev_sample["data"]:
        
        for context in wiki_article["paragraphs"]:
            
            for qas in context["qas"]:
                # print(qas)
                for i, ans in enumerate(qas["answers"]):
                    record_dict = {}
                    record_dict["title"] = wiki_article["title"]
                    record_dict["context"] = context["context"]
                    record_dict["question"] = qas["question"]
                    record_dict["id"] = qas["id"]
                    record_dict["answers"] = qas["answers"][i]["text"]
                    record_dict["answer_start"] = qas["answers"][i]["answer_start"]
                    record_dict["answer_type"] = qas["answers"][i]["answer_type"]
                    dev_sample_dict_list.append(record_dict)
    
    
    dev_sample_df = pd.DataFrame.from_records(dev_sample_dict_list)
    dev_df_list.append(dev_sample_df)

dev_df = pd.concat(dev_df_list, axis=0)
dev_df.shape
```
